```python
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris() #loading Iris
```

```python
digits = datasets.load_digits() #loading digits
```

```python
diabetes = datasets.load_diabetes() #loading diabetes
```

```python
iris_df = pd.DataFrame(
    iris.data,
    columns = iris.feature_names
)
```

```python
digits_df = pd.DataFrame(
    digits.data,
    columns = digits.feature_names
)
```

```python
diabetes_df = pd.DataFrame(
    diabetes.data,
    columns = diabetes.feature_names
)
```

```python
print("Datasets Loaded.")
```

```python
# Setting Target
iris_df['target'] = pd.Series(iris.target)
iris_df['target_names'] = iris_df['target'].apply(lambda x : iris.target_names[x])

digits_df['target'] = pd.Series(digits.target)
digits_df['target_names'] = digits_df['target'].apply(lambda x : digits.target_names[x])

diabetes_df['target'] = pd.Series(diabetes.target)
# diabetes_df['target_names'] = diabetes_df['target'].apply(lambda x : diabetes.target_names[x])
```

```python
print("Description of Each Dataset")

print("Iris: ")
print(iris.DESCR)

print("\n")

print("Digits: ")
print(digits.DESCR)

print("\n")

print("Diabetes: ")
print(diabetes.DESCR)
```

```python
from sklearn.model_selection import train_test_split
```

```python
iris_train, iris_test = train_test_split(iris_df, test_size = 0.2, random_state = 30)
```

```python
digits_train, digits_test = train_test_split(digits_df, test_size = 0.2, random_state = 30)
```

```python
diabetes_train, diabetes_test = train_test_split(diabetes_df, test_size = 0.2, random_state = 30)
```

```python
def print_details(df, train, test, name):
  print(name)
  print(f"Number of records in Dataset: {df.shape[0]}")
  print(f"Number of records in train dataset: {train.shape[0]}")
  print(f"Number of records in test dataset: {test.shape[0]}")
```

```python
print_details(iris_df, iris_train, iris_test, "Iris")
```

```python
print_details(digits_df, digits_train, digits_test, "Digits")
```

```python
print_details(diabetes_df, diabetes_train, diabetes_test, "Diabetes")
```
