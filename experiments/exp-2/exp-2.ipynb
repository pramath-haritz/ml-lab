```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data)
label=pd.DataFrame(iris.target)

X_train, X_test, y_train, y_test = train_test_split(df, label, test_size = 0.3, shuffle = True)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.3, shuffle = True)

print(f"Dataframe Rows: {df.shape[0]}")
print(f"Label Rows: {label.shape[0]}")

print(f"Input Train Rows: {X_train.shape[0]}")
print(f"Output Train Rows: {y_train.shape[0]}")

print(f"Input Test Rows: {X_test.shape[0]}")
print(f"Output Test Rows: {y_test.shape[0]}")

print(f"Input Validation Rows: {X_val.shape[0]}")
print(f"Output Validation Rows: {y_val.shape[0]}")

print()

print("Train First 5 Rows:")
print(X_train.head(3))
print("Train Random 5 Rows:")
print(X_train.sample(3))

print("Validation First 5 Rows:")
print(X_val.head(3))
print("Validation Random 5 Rows:")
print(X_val.sample(3))

print("Test First 5 Rows:")
print(X_test.head(3))
print("Test Random 5 Rows:")
print(X_test.sample(3))
```
